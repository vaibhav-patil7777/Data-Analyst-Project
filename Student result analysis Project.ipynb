# Impact of Parents Eductaion on Students Score

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("relation between students score on parents education")

plt.show()
```

```python
#From the above chart we have conclude the eduction of parents good impact on students performance and their marks
```

# Impact of Parents Marital status on Students Score

```python
gb2=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb2)
```

```python
sns.heatmap(gb2,annot=True)
plt.title("relation between students score on parents marital status")
plt.show()
```

```python
#from the above chart we conclude their is no negligible impact on students performance and marks compared to parents eduction
```

#                       Distribustion of Ethnic Group

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
l=["groupA","groupB","groupC","groupD","groupE"]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Groups")
```
